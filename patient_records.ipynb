```python
import json
import os 
from pprint import pprint
import glob
```

```python
working_dir_path = os.path.join(os.path.abspath(''), 'Patients_records')

json_dir_path = os.path.join(working_dir_path, 'fhir')

output_path1 = os.path.join(working_dir_path, 'diagnosis_dict.json')
output_path2 = os.path.join(working_dir_path, 'patient_dict.json')
output_path3 = os.path.join(working_dir_path, 'goals_dict.json')
```

```python
i = 0
clear = 0
diagnosis_dict = {}
goals_dict = {}
patient_dict = {}

for filename in glob.iglob(os.path.join(json_dir_path,'*.json')):
    parsed_json = json.loads(open(filename).read())
    
    diagnosis = []
    goals = []

    for entry in parsed_json.get("entry"):
        subentry = entry.get("resource")
        if subentry.get("resourceType") == "Condition":
            diagnosis.append(subentry.get("code").get("coding")[0].get("code"))
        if (subentry.get("resourceType") == "Goal")&(subentry.get("status") == "in-progress"):
            goals.append(subentry.get("description").get("text"))
        if subentry.get("resourceType") == "Patient":
            patient_id = subentry.get("identifier")[0].get("value")
            #print(patient_id)

    diagnosis = set(diagnosis)
    goals = set(goals)
    
    if len(diagnosis) == 0:
        clear += 1

    for disease in diagnosis:
        diagnosis_dict.setdefault(disease, []).append(patient_id)
    
    patient_dict.setdefault(patient_id, []).extend(diagnosis)
    goals_dict.setdefault(patient_id, []).extend(goals)
    
    i+= 1
    if i%50 == 0:
        print(i)

print(i)
print(clear)
```
